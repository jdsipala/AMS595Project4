```python
import numpy as np
```

```python
# part 1
P= np.random.rand(5,5)
# for each row 
for i in range(5):
    row_sum = 0
    # for each observation in the row
    for j in range(5):
        row_sum += P[i,j]
    # transform so the row sums to 1
    for j in range(5):
        P[i,j] = P[i,j] / row_sum
```

```python
# part 2
p=np.random.rand(5)
#initialize sum
total = 0
# find total of all 5 elements
for i in range(5):
    total += p[i]
# divide by total so that sum = 1
for i in range(5):
    p[i] = p[i] / total
```

```python
#part 2 continued
# transition rule: phat = P.T * p applied 50 times
for k in range(50):
    p = np.dot(P.T,p)
print(p.sum())
```

```python
# part 3
# use linalg to derive eigen vals / eigen vecs
eigenvalues, eigenvectors = np.linalg.eig(P.T)

closest_val = None
closest_dist = 1000
# get index of closest eigenvalue to 1
for i in range(len(eigenvalues)):
    diff = abs(eigenvalues[i] - 1)
    if diff < closest_dist:
        closest_dist = diff
        closest_val = i

# get the eigenvector that contains the closest eigenvalue to 1
v = eigenvectors[:, closest_val]

# if sum of eigenvector is negative make it positive
if v.sum() < 0:
    v = -v

# scale the vector so the sum = 1
v = v / v.sum()

print("Sum of v ", v.sum())
print("Stationary distribution is ", v)
```

```python
# part 4
cwisediff = np.abs(p-v)

if np.all(cwisediff < 1e-5):
    print("all components within 10^-5")
else:
    print("not all within 10^-5")
```
