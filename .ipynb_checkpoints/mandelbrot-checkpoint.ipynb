```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make 50 by 50 grid of real (x) and imaginary (y)
x, y = np.mgrid[-2:1:50j, -1.5:1.5:50j]

# turn coordinates into c = x + i (50 x 50) 
c = x + 1j * y
```

```python
# initialize threshold and N_max
threshold = 50
N_max = 200
```

```python
# iterate zn+1 = z^2_n + c and create mask 
# z starts as all zeros, same shape as c
z = np.zeros(c.shape, dtype = np.complex64)
# 50 by 50 all true
mask = np.ones(c.shape, dtype= bool)
```

```python
# iterate the mandelbrot equation 200 times
for n in range(N_max):
    # update z using z = z^2 + c
    for i in range(z.shape[0]):
        for j in range(z.shape[1]):
            if mask[i,j] == True:
                z[i,j] = z[i,j] * z[i,j] + c[i,j]
    #after each loop, update mask
    mask = np.abs(z) <= threshold
```

```python
plt.imshow(mask.T, extent =[-2,1,-1.5,1.5])
plt.gray()
plt.show()
```
